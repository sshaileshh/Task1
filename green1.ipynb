```python
import selenium
from selenium import webdriver as wb
webD=wb.Chrome('chromedriver.exe')
webD.get('https://www.matchesfashion.com/intl/mens')
```

```python
clickObj=webD.find_element_by_xpath('//*[@id="nav_main"]/ul/li[4]/span/a')
clickObj.click()
```

```python
x=webD.find_element_by_xpath('//*[@id="page"]/div[4]/div[1]/div[4]/div[9]/ul')
```

```python
productInfolist=webD.find_elements_by_class_name('lister__item')
```

```python
listoflinks=[]
for el in productInfolist:
    pp1=el.find_elements_by_tag_name('div')[0]
    pp2=pp1.find_element_by_tag_name('a')
    listoflinks.append(pp2.get_property('href'))
```

```python
listoflinks
```

```python
from tqdm import tqdm
alldetails=[]
for i in tqdm(listoflinks):
    webD.get(i)
name=webD.find_element_by_xpath('//*[@id="page"]/div[4]/div[1]/div[4]/div[9]/ul/li[1]/div/a/div[2]').text
brand=webD.find_element_by_xpath('//*[@id="page"]/div[4]/div[1]/div[4]/div[9]/ul/li[1]/div/a/div[1]').text
price=webD.find_element_by_xpath('//*[@id="page"]/div[4]/div[1]/div[4]/div[9]/ul/li[1]/div/a/div[3]/span').text
tempJ={'name':name,
     'brand':brand,
     'price':price,
      'productLink':i}
alldetails.append(tempJ)
```
